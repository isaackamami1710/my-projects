```python
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
def calc_vif(x):
    vif=pd.DataFrame()
    vif['column']=x.columns
    vif['vif']=[variance_inflation_factor(x.values,i) for i in range(x.shape[1])]
    return vif
```

```python
calc_vif(credit)
```

```python
X=credit.iloc[:,:-1]
y=credit.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.45,train_size=0.55,random_state=90)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
clf=DecisionTreeClassifier(criterion='entropy',max_depth=20,splitter='best',random_state=102,min_samples_split=5,ccp_alpha=0.01)
clf.fit(X_train,y_train)
```

```python
print('the model test score',clf.score(X_test,y_test))
print('the model train score',clf.score(X_train,y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model=RandomForestClassifier(n_estimators=10,max_depth=5,n_jobs=-1,random_state=105,criterion='entropy')
model.fit(X_train,y_train)
```

```python
print('model test_score',model.score(X_test,y_test))
print('model train_score',model.score(X_train,y_train))
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
pred=model.predict(X_test)
```

```python
print("Accuracy score",accuracy_score(y_test,pred))
```

```python
metrix=confusion_matrix(y_test,pred)
metrix
```

```python
from sklearn.naive_bayes import BernoulliNB
```

```python
model=BernoulliNB(binarize=0.2,alpha=0.5,fit_prior=True,class_prior=None)
model.fit(X_train,y_train)
```

```python
print('model test score',model.score(X_test,y_test))
print('model train score',model.score(X_train,y_train))
```
